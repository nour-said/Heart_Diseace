```python
data = df.copy()
data
```

```python
num_data=[]
for i in data.columns:
    if data[i].dtype==float or  data[i].dtype==int :
        num_data.append(i)
print(num_data)
```

```python
from scipy.stats import zscore
import numpy as np
Z_score= zscore(data[num_data])
Z_score
```

```python
Abs_scores = np.abs(Z_score)
Abs_scores
```

```python
df = df[df.columns].replace({'Yes' : 1, 'No': 0, 'Male': 1, 'Female' :0, 'Poor' :1, 'Fair' :0, 'Good': 2, 'Very good' :3, 'Excellent' :4, 'No, borderline diabetes':0, 'Yes (during pregnancy)':1})
print(df)
```

```python
df['BMI'] = Abs_scores['BMI'].values
df['PhysicalHealth'] = Abs_scores['PhysicalHealth'].values
df['MentalHealth'] = Abs_scores['MentalHealth'].values
df['SleepTime'] = Abs_scores['SleepTime'].values

df
```

```python
from sklearn.preprocessing import LabelEncoder
le =  LabelEncoder()
cate = []
cate.append(df.AgeCategory)
df['AgeCategory']=le.fit_transform(df['AgeCategory'])
# cate
df
```

```python
y = df['HeartDisease']
X = df.drop(columns = ['HeartDisease'])
```

```python
from sklearn.preprocessing import MinMaxScaler

# fit scaler on training data
norm = MinMaxScaler().fit(X)

# transform training data
X = norm.transform(X)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import f_classif

test = SelectKBest(score_func=chi2, k=8)

fit = test.fit(X, y)
np.set_printoptions(precision=6)
print(fit.scores_)

X = fit.transform(X)

print(X[0:5,:])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state= 42,  shuffle = True)
```

```python
log_reg = LogisticRegression(max_iter=2000)
log_reg.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

predictions = log_reg.predict(X_train)
percentage = log_reg.score(X_train,y_train)

res=confusion_matrix(y_train,predictions)
print("confusion matrix")
print(res)

print(f"Accuracy={percentage*100}%")
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

predictions = log_reg.predict(X_test)
percentage = log_reg.score(X_test,y_test)

res=confusion_matrix(y_test,predictions)
print("confusion matrix")
print(res)

print(f"Accuracy={percentage*100}%")
```
